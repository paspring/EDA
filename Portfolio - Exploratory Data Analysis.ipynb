#### Cleaning Variables

In the `nbrnaliv` variable is described the types of born alive pregnancies, but one of the respondent refused to answer and that case was labeled with a 8 number, because thats not interesting for us we can replace that value for an nan.

<img src="img/nbrnaliv.png" width=600>

```python
# Replace the value 8 with NaN
nsfg['nbrnaliv'].replace(8, np.nan, inplace=True)

# Print the values and their frequencies
print(nsfg['nbrnaliv'].value_counts())
```

#### Manipulating some variables.

<img src="img/agecon.png" width=300>

```python
nsfg.head()
```

```python
# Select the columns and divide by 100
agecon = nsfg['agecon'] / 100
agepreg = nsfg['agepreg'] / 100

# Compute the difference
preg_length = agepreg - agecon

# Compute summary statistics
print(preg_length.describe())
```

#### Birth Weight Variable.

This variable have enconding that change the result of our distributions.

<img src="img/weight.png" width=500>

Wee need to fill the 98 and 99 encoding with nan.

```python
nsfg[nsfg.birthwgt_lb1>=98].shape, nsfg[nsfg.birthwgt_oz1>=98].shape
```

```python
# Clean the weight variables
pounds = nsfg['birthwgt_lb1'].replace([98, 99], np.nan)
ounces = nsfg['birthwgt_oz1'].replace([98, 99], np.nan)

# Compute total birth weight
birth_weight = pounds + ounces/16
nsfg["birth_weight"]=birth_weight
nsfg.head()
```

#### Difference of weight between single and multiple born.

```python
# Filter full-term babies
full_term = nsfg['prglngth'] >= 37

# Filter single births
single = nsfg['nbrnaliv']==1

# Compute birth weight for single full-term babies
single_full_term_weight = nsfg.birth_weight[full_term & single]
print('Single full-term mean:', single_full_term_weight.mean())

# Compute birth weight for multiple full-term babies
mult_full_term_weight = nsfg.birth_weight[full_term & ~single]
print('Multiple full-term mean:', mult_full_term_weight.mean())
```

### Modeling distributions

#### Distribution of income

In many datasets, the distribution of income is approximately lognormal, which means that the logarithms of the incomes fit a normal distribution. We'll see whether that's true for the GSS data. 

We'll compute the mean and standard deviation of the log of incomes using NumPy's np.log10() function.

```python
# Extract realinc and compute its log
income = gss['realinc']
log_income = np.log10(income)

# Compute mean and standard deviation
mean = np.mean(log_income)
std = np.std(log_income)
print(mean, std)

# Make a norm object
from scipy.stats import norm
dist = norm(mean,std)
```

#### Comparing CDFs

To see whether the distribution of income is well modeled by a lognormal distribution, we'll compare the CDF of the logarithm of the data to a normal distribution with the same mean and standard deviation. 



```python
# Evaluate the model CDF
xs = np.linspace(2, 5.5)
ys = norm(mean, std).cdf(xs)

# Plot the model CDF
plt.clf()
plt.plot(xs, ys, ':r', label="LogNorm Model")

# Create and plot the Cdf of log_income
Cdf.from_seq(log_income).plot(label="Income Distribution")
    
# Label the axes
plt.xlabel('log10 of realinc')
plt.ylabel('CDF')
plt.legend()
plt.show()
print("The lognormal model is a pretty good fit for the data, but clearly not a perfect match. ")
```

#### Comparing PDFs

In the previous exercise, we used CDFs to see if the distribution of income is lognormal. We can make the same comparison using a PDF and KDE. 


```python
# Evaluate the normal PDF
xs = np.linspace(2, 5.5)
ys = norm(mean,std).pdf(xs)

# Plot the model PDF
plt.clf()
plt.plot(xs, ys, color='gray', label='General Model Distribution')

# Plot the data KDE
sns.kdeplot(log_income, label="Income")

# Label the axes
plt.xlabel('log10 of realinc')
plt.ylabel('PDF')
plt.show()
```

#### Using StatsModels

Let's run the same regression using SciPy and StatsModels, and confirm we get the same results.

- Compute the regression of '_VEGESU1' as a function of 'INCOME2' using SciPy's linregress().
- Compute the regression of '_VEGESU1' as a function of 'INCOME2' using StatsModels' smf.ols()

```python
from scipy.stats import linregress
import statsmodels.formula.api as smf

# Run regression with linregress
subset = brfss.dropna(subset=['INCOME2', '_VEGESU1'])
xs = subset['INCOME2']
ys = subset['_VEGESU1']
res = linregress(xs, ys)
print(res)

# Run regression with StatsModels
results = smf.ols('_VEGESU1 ~ INCOME2', data=brfss).fit()
print(results.params)
```

#### Sex and income

Using `C(sex)` we can add sex as a categorical variable on the model.

```python
gss['educ2']=gss.educ**2
gss['age2']=gss.age**2
```

```python
formula = 'realinc ~ educ + educ2 + age + age2 + C(sex)'
results = smf.ols(formula, data=gss).fit()
results.params
```

`C(sex)[T.2] = -4156.113865` means a base difference of income between male (1) and females (2).

#### Gun control laws between sex.

Let's work with a boolean variable, if we change $gunlaw = 2$ to 0 we have a boolean variable to add in our model.

```python
gss['gunlaw'].replace([2], [0], inplace=True)
print(gss['gunlaw'].value_counts())
print('Here we can see that 3/4 of the respondents is in favor of gun control laws.')
```

```python
formula = 'gunlaw ~ age + age2 + educ + educ2 + C(sex)'
results = smf.logit(formula, data=gss).fit()
results.params
```
